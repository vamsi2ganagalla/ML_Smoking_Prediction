```python
numeric_cols = train_df.select_dtypes(include=[int, float]).columns.tolist()
numeric_cols.remove("smoking")

categorical_cols = [col for col in train_df.columns if train_df[col].nunique() <= 10 and col != "smoking"]

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(y='smoking', data=train_df, palette='coolwarm')
plt.title("Target Variable Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(train_df.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_cols = train_df.select_dtypes(include=['int64','float64']).columns.tolist()
numeric_cols.remove("smoking")

# Calculate the number of rows and columns for subplots
n_cols = 5  # Adjust as needed
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # Ceiling division

plt.figure(figsize=(n_cols * 4, n_rows * 3))

for i, col in enumerate(numeric_cols):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.boxplot(x=train_df[col])
    plt.title(col)
    plt.xlabel('') # Remove x-axis label for cleaner look

plt.tight_layout()
plt.suptitle("Distribution of Numeric Features", y=1.02, fontsize=16)
plt.show()
```
